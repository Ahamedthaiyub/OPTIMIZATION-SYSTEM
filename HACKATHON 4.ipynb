### Add filter columns for all the places where the visitor is 'IDLE'

```python
# filter for where first dwell is 'idle'
df2['idle0'] = (df2['Space1'] == ' OTHER')
df2['idle0'] = df2['idle0']*1

df2['idle1'] = (df2['Space2'] == ' OTHER')
df2['idle1'] = df2['idle1']*1

df2['idle2'] = (df2['Space3'] == ' OTHER')
df2['idle2'] = df2['idle2']*1

df2['idle3'] = (df2['Space4'] == ' OTHER')
df2['idle3'] = df2['idle3']*1
# -------------- breaker --- so I don't go cross-eyed
df2['idle4'] = (df2['Space5'] == ' OTHER')
df2['idle4'] = df2['idle4']*1

df2['idle5'] = (df2['Space6'] == ' OTHER')
df2['idle5'] = df2['idle5']*1

df2['idle6'] = (df2['Space7'] == ' OTHER')
df2['idle6'] = df2['idle6']*1

df2['idle7'] = (df2['Space8'] == ' OTHER')
df2['idle7'] = df2['idle7']*1
```

### Binning times

```python
import dateutil
df2['dateok'] = df2['Date'].apply(dateutil.parser.parse, dayfirst=True)
```

```python
df2.head()
```

```python
df2.dtypes
```

```python
import datetime
df2['Date'] = [datetime.datetime.date(d) for d in df2['dateok']]# extracting date from timestamp
df2['Time'] = [datetime.datetime.time(d) for d in df2['dateok']]# extracting time from timestamp
```

```python
df2.head()
```

```python
df2['hod'] = [t.hour for t in df2.Time]
df2.head()
```

```python
df2['interval_of_day'] = pd.cut(df2['hod'], bins=4, labels=['morning', 'mid-day', 'evening', 'night'])
```
